```python
# 目标：查看T飞行和T-1飞行是否存在闭环关系
# 链式遍历：[1,2], [2,3], [3,]
# 情况：
# 1、[1]
# 2、[1,2], [2,3]
# 3、[1,2], [2,3,], [3,]
# 飞行路径：
# a-b        -> 开区间，近期存在b-a的可能性高
# a-b、b-a   -> 闭区间，近期有行程可能性低
# a-b、b-c   -> 开区间，存在c-a的可能性高
```

```python
class PsrFlyRecord:

    def __init__(self, records=None):
        self.record = records
        self.udlist = []
        self.up_locations = []
        self.dis_locations = []
        self.udlines = []
        self.airlines = []
        self.up_counter = None
        self.dis_counter = None
        self.udlines_counter = None
        self.airlines_counter = None

    def input_udlist(self, udlist):
        """
            udlist: [[up, dis],[up, dis], ...]
        """
        self.udlist = udlist
        for ud in self.udlist:
            u = ud[0]
            d = ud[1]
            copied_ud = ud.copy()
            copied_ud.sort()
        
            self.up_locations.append(u)
            self.dis_locations.append(d)
            self.udlines.append("-".join(ud))
            self.airlines.append("=".join(copied_ud))

    def most_freq_up(self, num=1):
        """ 获取最多的出发地
        """
        self.up_counter = Counter(up_locations)
        return dict(self.up_counter.most_common(num))

    def most_freq_dis(self, num=1):
        """ 获取最多的目的地
        """
        self.dis_counter = Counter(dis_locations)
        return dict(self.dis_counter.most_common(num))

    def most_freq_udline(self, num=1):
        """ 获取最多的航线
        TODO: HAK-TYN, TYN-HAK ->  HAK=TYN
        """
        self.udlines_counter = Counter(self.udlines)
        return dict(self.udlines_counter.most_common(num))

    def most_freq_airline(self, num=1):
        self.airlines_counter = Counter(self.airlines)
        return dict(self.airlines_counter.most_common(num))
```

```python
psrfly = PsrFlyRecord()
psrfly.input_udlist(ud_list)
```

```python
up = psrfly.most_freq_up(3)
dis = psrfly.most_freq_dis(3)
line = psrfly.most_freq_udline(2)
airline = psrfly.most_freq_airline(2)
print(up, dis, line, airline)
```
